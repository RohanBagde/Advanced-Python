# class

```python
#syntax for class
#def name of class(self,list of formal parameter if any)
class studentex:pass #class is programmer define datatype
                                                         #(without class level data member)


#main program
s1=studentex()
#place the data for s1
s1.stno=20
s1.name="rohan"
s1.marks=22.22
s1.addr="nagpur"
#create another object
s2=studentex()
#place the data for s2
s2.stno=30
s2.name="goldberg"
s2.marks=33.22
s2.addr="newyork"
print("-"*30)
print("first student data")
print("-"*30)
print("id of student1={}".format(id(s1)))
print("student number={}".format(s1.stno))
print("student name={}".format(s1.name))
print("student marks={}".format(s1.marks))
print("student addr={}".format(s1.addr))
print("-"*30)
print("second student data")
print("id of stuudent2={}".format(id(s2)))
print("-"*30)
print("student number={}".format(s2.stno))
print("student name={}".format(s2.name))
print("student marks={}".format(s2.marks))
print("student addr={}".format(s2.addr))

```

```python
#syntax for class
#def name of class(self,list of formal parameter if any)
class studentex: #class is programmer define datatype  #("delete pass")
    csr="PYTHON"                                                          #(with class level data member)


#main program
s1=studentex()
#place the data for s1
s1.stno=20
s1.name="rohan"
s1.marks=22.22
s1.addr="nagpur"
#create another object
s2=studentex()
#place the data for s2
s2.stno=30
s2.name="goldberg"
s2.marks=33.22
s2.addr="newyork"
print("-"*30)
print("first student data")
print("-"*30)
print("id of student1={}".format(id(s1)))
print("student number={}".format(s1.stno))
print("student name={}".format(s1.name))
print("student marks={}".format(s1.marks))
print("student addr={}".format(s1.addr))
print("first student course={}".format(studentex.csr)) #automatically value added just lie defalt parameter
print("-"*30)
print("second student data")
print("id of stuudent2={}".format(id(s2)))
print("-"*30)
print("student number={}".format(s2.stno))
print("student name={}".format(s2.name))
print("student marks={}".format(s2.marks))
print("student addr={}".format(s2.addr))
print("second student course={}".format(studentex.csr))  #use class name thats why called class level data type
```
