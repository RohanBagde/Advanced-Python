# Synchronization Concept in Multi-Threading

=> When multiple threads are operating /working on same object / resource / Method then by default we get dead lock result / race condition (wrong result) / non-thread safety result.
=>To overcome this dead lock result , industry is recommeded to apply synchronization / locking concept.
=>The advantage of synchronization /  locking  concept is that to provide Thread Safety Result (Consistant /accurate  result)
=>In python we can obtain the synchronization by using locking and un-locking concept.

=>Steps for implementing Synchroization in python
-----------------------------------------------------------------------------
1) obtain / create  an object of Lock class of threading module
	Example:      l=threading.Lock()
    
2) To lock the current object / method  we must use acquire() 
		Example:-    l.acquire()
        
3) To unlock or release the object/method , we can use release()
		l.release()

```python
#nonsync1-----------functional aproach
from threading import *
def table(n):
    print("-"*50)
    print("mul table for={}".format(n))
    print("-"*50)
    for i in range(1,11):
        print("\t{} X {}={}".format(n,i,n*i))
    else:
        print("-"*50)
#main program
t1=Thread(target=table,args=(5,))
t2=Thread(target=table,args=(6,))
t3=Thread(target=table,args=(7,))
t4=Thread(target=table,args=(8,))
t5=Thread(target=table,args=(9,))
#start the threads
t1.start()
t2.start()
t3.start()
t4.start()            #unaccepted result or deadlock result or nonsync or diff order
t5.start()
```

# Locking concept

```python
#sync1
#locking facility 
from threading import *
import time
l=Lock()  #global variable--------create an object of lock
def table(n):
    l.acquire()  #get the lock
    print("-"*50)
    print("mul table for={}".format(n))
    print("-"*50)
    for i in range(1,11):
        print("\t{} X {}={}".format(n,i,n*i))
        time.sleep(1)
    else:
        print("-"*50)
        l.release()   #release the lock
#main program
t1=Thread(target=table,args=(5,))
t2=Thread(target=table,args=(6,))
t3=Thread(target=table,args=(7,))
t4=Thread(target=table,args=(8,))
t5=Thread(target=table,args=(9,))
#start the threads
t1.start()
t2.start()
t3.start()
t4.start()            
t5.start()
```

```python
#synchex2.py-----OOPS principles
from threading import *
import time
class MulTable1:
	def __init__(self):
		self.l=Lock()	# create an object of lock
	def  table(self,n):
		self.l.acquire()   # get the the lock of table() of mt object
		print("-"*50)
		print("{} obtains lock on table()".format(current_thread().name))
		print("Mul table for {}".format(n))
		print("-"*50)
		for i in range(1,11):
			print("\t{} x {}={}".format(n,i,n*i))
			time.sleep(1)
		else:
			print("-"*50)
		self.l.release()   # release the the lock of table() of mt object
#main program
mt=MulTable1()
t1=Thread(target=mt.table,args=(4,))
t2=Thread(target=mt.table,args=(14,))
t3=Thread(target=mt.table,args=(19,))
t4=Thread(target=mt.table,args=(3,))
t1.name="Rossum"   # user-friendly names to the threads
t2.name="Gosling"
t3.name="Ritche"
t4.name='Travis'
t1.start()
t2.start()
t3.start()
t4.start()
```
